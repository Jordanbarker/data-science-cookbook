As race day approaches, pay particular attention to eating plenty of carbohydrates and drinking fluids. Aim for a diet consisting of about 55-70 per cent carbohydrate. It should be fairly low in fat, easy to digest, and with familiar food selections.

- Carbs are important
- You should consume foods your body is already used to eating.

Examples:
- white rice
- sweet potatoes 
- pasta with marinara

---

Total running volume (weekly mileage) is the single most important factor.
- Increase weekly mileage by no more than 5–10% each week
- Cutback weekly mileage every 3–4 weeks by 15–20% 

- 80% of training at low intensity and ~20% at moderate to high intensity
- hard day, easy day pattern

The final 2–3 weeks before the marathon are a tapering period. Gradually reduce mileage by ~40–60% by race week. For example, if the peak was 50 miles, drop to ~35 miles two weeks out and ~20–25 miles in the last week. Intensity is maintained with a few short faster workouts.​

---

Workout Types:
- Long Slow Distance: 1–2 minutes per mile slower than goal marathon pace.
	- 2x per week. 
	- Cap to 18–20 miles (<3-3.5 hours).

- Strides: accelerations, typically 15–30 second, done at the end of easy runs.
	- 2x–3x per week, 4–6 strides
	- Meant to be swift and smooth, not exhausting.

- Tempos: Sustained effort at lactate threshold pace
	- 1x per week or every 2 weeks

- Intervals (VO2max Workouts): Repeats at 5K/10K pace
	- 1x per week
	- Add easy jog recoveries at the end.
	- Examples: 
		5 x 1000 m at 5K pace, 
		8 x 800 m at 10K pace, 
		ladder workouts like 1200-1000-800-600-400 m at progressively faster paces

- Hill workouts: 1–3 minutes at a hard effort up a moderate incline
	- develops leg strength

Training Intensity Distributions (TID):
- Polarized model: 80% in z1, remaining 20% in z3.
- Pyramidal model: 70–80% in z1, remaining 20–30% in z2 and z3. It is a pyramid because z1 > z2 > z3
- Threshold model: >35% in z2, remaining in z1 and z3. 

Pyramidal seems to be the most commonly used, but research is inconclusive on the optimal split. Regular training in zone 2 will be important for athletes whose competition duration exceeds 30 minutes.

There are also individual variances. Are you a recreational or expert runner? Are you improving or plateauing with what you are currently doing? Are you recovering well enough to run hard for the high intensity workouts?

Elite athletes rarely perform polarized training and there is limited to-no evidence that this is optimal for endurance athletes.

Debate:

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


def generate_weekly_mileage(
    start_miles: float,
    num_weeks: int,
    increase_pct: float = 0.07,
    cutback_pct: float = 0.15,
    cycle_length: int = 4,
    taper_weeks: int = 3,
    taper_mid_drop: float = 0.25,
    taper_final_drop: float = 0.50,
) -> list[float]:
    """
    Generates a weekly running mileage plan that includes progressive loading,
    periodic cutbacks, and a final taper before race week.

    Args:
        start_miles: Mileage for week 1.
        num_weeks: Length of the plan, including taper weeks.
        increase_pct: Weekly fractional increase outside cutbacks/taper.
        cutback_pct: Fractional drop on scheduled cutback weeks.
        cycle_length: Number of weeks between cutbacks (e.g., 4 → cutback every 4th week).
        taper_weeks: Length of the taper.
        taper_mid_drop: Fractional drop from peak mileage for the first taper week (e.g., 0.30 → 30% drop).
        taper_final_drop: Fractional drop from peak mileage for race week (e.g., 0.55 → 55% drop).

    Returns:
        A list containing mileage for each week (rounded to two decimals).

    Raises:
        ValueError: If num_weeks is not greater than taper_weeks.
    """
    if num_weeks <= taper_weeks:
        raise ValueError("num_weeks must exceed taper_weeks")

    mileage: list[float] = [start_miles]

    # Build mileage until taper phase begins.
    for week in range(2, num_weeks - taper_weeks + 1):
        if (week - 1) % cycle_length == 0:
            new = mileage[-1] * (1 - cutback_pct)
        else:
            new = mileage[-1] * (1 + increase_pct)
        mileage.append(round(new, 2))

    peak = max(mileage)

    # Construct taper—linear drop from mid‑taper to final‑taper targets.
    taper_factors = np.linspace(
        1 - taper_mid_drop, 1 - taper_final_drop, taper_weeks
    )
    for factor in taper_factors:
        mileage.append(round(peak * factor, 2))

    return mileage


# Params
start_miles = 25.0
weeks = 20
taper_weeks = 3

mileage_df = pd.DataFrame(
    {
        "5%": generate_weekly_mileage(start_miles, weeks, increase_pct=0.05, taper_weeks=3),
        "7%": generate_weekly_mileage(start_miles, weeks, increase_pct=0.07, taper_weeks=3),
        "10%": generate_weekly_mileage(start_miles, weeks, increase_pct=0.10, taper_weeks=3),
    }
)

plt.figure(figsize=(10, 6))
for col in mileage_df.columns:
    plt.plot(range(1, weeks + 1), mileage_df[col], label=col)
    plt.scatter(range(1, weeks + 1), mileage_df[col], marker=".")

plt.axvline(x=weeks - taper_weeks, color='gray', linestyle='--', linewidth=2, label="Taper Start")

plt.xlabel("Week")
plt.ylabel("Mileage")
plt.title("Weekly Running Mileage Plans (with 3-Week Taper)")
plt.legend(title="Increase Rate")
plt.grid(True, alpha=0.4)
plt.tight_layout()
plt.show()
```
